```python
import os

import pandas as pd
from PIL import Image
```

```python
data_path = '/group-volume/orc_srr/multimodal/iceblood/datasets/main/TRAIN/RTSD/'
df_path = "/group-volume/orc_srr/multimodal/iceblood/russian/full-gt.csv"
```

```python
df = pd.read_csv(df_path)
```

```python
df.head(10)
```

```python
df['path'] = df['filename'].apply(lambda x: os.path.join('rtsd-frames', x))

df['xtl'] = df['x_from'] 
df['ytl'] = df['y_from']
df['xbr'] = df['x_from'] + df['width']
df['ybr'] = df['y_from'] + df['height']

df['class'] = df['sign_class'].apply(lambda x: x.replace('_', '.'))
df['class'] = df['class'].apply(lambda x: x.split('.n')[0])
df['class'] = df['class'].apply(lambda x: x.split('.r')[0])
df['class'] = df['class'].apply(lambda x: '.'.join(x.split('.')[:3]))

df['temporary'] = None
df['occluded'] = None
```

```python
df = df.drop(['filename', 'x_from', 'y_from', 
              'width', 'height', 'sign_class', 'sign_id'], axis=1)
```

```python
df.to_csv(os.path.join(data_path, 'rtds_ann.csv'), index=False)
```
